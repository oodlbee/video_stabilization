```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from stabilizer import Stabilizator

input_video_path = 'data//day_light_traffic.mov'
```

```python
scale = 0.5
step = 100

stabilizer = Stabilizator(input_video_path, scale, step)
```

```python
stabilizer.frames_number
```

```python
stabilizer.calculate_warp_stack()
```

```python
i, j = 0, 2
plt.scatter(np.arange(len(stabilizer.warp_stack)), stabilizer.warp_stack[:,i,j], label='X Velocity')
plt.plot(np.arange(len(stabilizer.warp_stack)), stabilizer.warp_stack[:,i,j])
plt.scatter(np.arange(len(stabilizer.warp_stack)), np.cumsum(stabilizer.warp_stack[:,i,j], axis=0), label='X Trajectory')
plt.plot(np.arange(len(stabilizer.warp_stack)), np.cumsum(stabilizer.warp_stack[:,i,j], axis=0))
plt.legend()
plt.xlabel('Frame')
```

```python
i, j = 1, 2
plt.scatter(np.arange(len(stabilizer.warp_stack)), stabilizer.warp_stack[:,i,j], label='X Velocity')
plt.plot(np.arange(len(stabilizer.warp_stack)), stabilizer.warp_stack[:,i,j])
plt.scatter(np.arange(len(stabilizer.warp_stack)), np.cumsum(stabilizer.warp_stack[:,i,j], axis=0), label='X Trajectory')
plt.plot(np.arange(len(stabilizer.warp_stack)), np.cumsum(stabilizer.warp_stack[:,i,j], axis=0))
plt.legend()
plt.xlabel('Frame')
```
